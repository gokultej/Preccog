```python
from nltk.tokenize import word_tokenize
tokens = word_tokenize(' '.join(words1))
```

```python
from itertools import combinations
word_pairs = list(combinations(tokens, 2))
```

```python
def word_similarity(word1, word2):
    synsets1 = wordnet.synsets(word1)
    synsets2 = wordnet.synsets(word2)
    if not synsets1 or not synsets2:
        return 0 
    synset1 = synsets1[0]
    synset2 = synsets2[0]
    similarity_score = synset1.wup_similarity(synset2)
    return similarity_score if similarity_score is not None else 0
```

```python
similarity_scores = [word_similarity(pair[0], pair[1]) for pair in word_pairs]
```

```python
average_similarity = sum(similarity_scores) / len(similarity_scores)
```

```python
print("Average Similarity Score:", average_similarity)
```

```python
sentences = []
similarity_scores = []
```

```python
for pair in word_pairs:
    sentence = f"{pair[0]} {pair[1]}"
    similarity = word_similarity(pair[0], pair[1])
    sentences.append(sentence)
    similarity_scores.append(similarity)
```

```python
for i, (sentence, score) in enumerate(zip(sentences, similarity_scores), 1):
    print(f"Pair {i}: {sentence}")
```
